```python
def welcome(user):
    return "Welcome " + user.capitalize()

def dashord(word,chars = None):
    print("\n")
    if chars:
        new_word = ""
        for j in word:
            if j in chars:
                new_word += j
            else:
                new_word += " _ "
        if "_" in new_word:
            print(new_word+"\n")
            print("A B C D E F G Ğ H I İ J\n K L M N O Ö P R S Ş T Z\n")
        else:
            print("Tebrikler! Bulduğunuz kelime:: " + new_word)
            return 1
    else:
        print("_ "*len(word)+"\n")
        print("A B C D E F G Ğ H I İ J\n K L M N O Ö P R S Ş T Z\n")
        


def Hangman_Game():
    import random as rnd
    user_health = 3
    
    words = ["hangman","kütüphane","aşı"]
    
    print("Welcome to hangman game\n")

    username = input("Name:: ")
    print("\n" +welcome(username))
    print("\n------------------Game Started------------------\n")
    
    random_word = rnd.choice(words)
    dashord(random_word)
    char_list = []
    while user_health > 0:
        print("Kalan can :: " + str(user_health))
        char = input("-> ")
        if char.lower() in random_word:
            char_list.append(char)
            check = dashord(random_word,char_list)
            if check:
                break
            else:
                dashord(random_word,char_list)
        else:
            user_health -= 1
    print("Oyun bitti!")
    
Hangman_Game()
```
